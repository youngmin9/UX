```python
#분석 진행할 날짜 지정
month_analy = int(input("분석 월 : "))
day_analy = int(input("분석 일 : "))

#Coffee history import
df_coffee_history = pd.read_csv('/content/drive/Shareddrives/Algorithm/카페인 데모데이터/Test coffee.csv', encoding='cp949')
df_coffee_history.columns = ['Name', 'Coffee_time', 'Coffee_dose']
df_coffee_history['Coffee_time'] = pd.to_datetime(df_coffee_history['Coffee_time'])

#Coffee history 데이터의 날짜 바꾸기
for i in range(len(df_coffee_history)):
  df_coffee_history.loc[i,'Coffee_time'] = df_coffee_history.loc[i, 'Coffee_time'].replace(month = month_analy, day = day_analy)

#초기 설문 조사 import
df_demo = pd.read_csv('/content/drive/Shareddrives/Algorithm/카페인 데모데이터/Test demo.csv', encoding='cp949')
df_demo.columns = ['Name', 'Weight', 'Pregnancy', 'OC', 'Smoking']

#부작용 설문 조사 import
df_effect = pd.read_csv('/content/drive/Shareddrives/Algorithm/카페인 데모데이터/Test result.csv', encoding='cp949')
df_effect.columns = ['Name', 'Adverse', 'Sleep_time', 'Sleep_adverse']
df_effect['Sleep_time'] = pd.to_datetime(df_effect['Sleep_time'])
for i in range(len(df_effect)):
  df_effect.loc[i,'Sleep_time'] = df_effect.loc[i,'Sleep_time'].replace(month = month_analy, day = day_analy)

#이름 자동으로 불러오기
people_list = df_coffee_history['Name'].to_list()
people_list = list(set(people_list))
print (people_list)
for person_name in people_list:
  person_file_name = "/content/drive/Shareddrives/Algorithm/카페인_1기/고객csv/HR_" + person_name + "_24" + format(month_analy, '02') + format(day_analy, '02') + ".csv"

  #Import df에서 Target의 데이터만 추출 (현재는 지정 -> 후에는 입력하면 자동화할 수 있게)
  df_demo_target = df_demo.loc[df_demo.Name == person_name, :]
  column_names = ["Weight", "Pregnancy", "OC", "Smoking"]
  df_demo_target.reset_index(inplace = True)

  #데이터의 각 값을 변수에 할당 - 변수명은 column 명과 동일
  for demo_category in column_names:
      exec("%s = %d" % (demo_category, df_demo_target.at[0,demo_category]))

  #개인별 ke 값 계산
  ke = 0.13857 * (1+1.036*Smoking) * (1-0.4*OC) * (1-Pregnancy/3)

  #설문 조사 기반 개인 식 산출 (t: 현재 시점)
  def single_dose(drink_time, dose, t):
      global single_caffeine_dose
      drink_time_hour = drink_time.hour + drink_time.minute / 60 + drink_time.second / 3600
      if t.day == drink_time.day + 1:
        t_hour = t.hour + t.minute / 60 + t.second / 3600 + 24
      else:
        t_hour = t.hour + t.minute / 60 + t.second / 3600
      single_caffeine_dose = dose * (198 / (138.6 - 7 * ke)) * (math.exp((-ke)*(t_hour - drink_time_hour)/(0.7)) - math.exp((-19.8)*(t_hour - drink_time_hour))) / Weight
      return single_caffeine_dose

  #분석 대상 지정
  df_coffee_history_person = df_coffee_history.loc[df_coffee_history.Name == person_name, :]
  df_coffee_history_person.reset_index(inplace = True)

  set_caffeine_dose = 0

  #특정 시점의 caffeine dose 계산
  def total_dose(t):
      df_coffee_history_use = df_coffee_history_person.loc[df_coffee_history_person.Coffee_time < t, :]
      global total_caffeine_dose
      global set_caffeine_dose
      for i in range (len(df_coffee_history_use)):
          set_caffeine_dose = set_caffeine_dose + single_dose(df_coffee_history_use.at[i,'Coffee_time'], df_coffee_history_use.at[i,'Coffee_dose'], t)
      total_caffeine_dose = set_caffeine_dose
      set_caffeine_dose = 0
      return total_caffeine_dose

  #1분 단위로 dose 계산해서 DF에 저장
  df_caffeine = pd.DataFrame(columns = ['Time', 'Dose'])

  #다음날 새벽 6시까지 caffeine 그래프 그릴 수 있도록 바꿔두었음!!
  for i in range(1800):
    if i < 1440:
      dose_time = datetime.strptime("2024-{}-{} {}:{}:00".format(month_analy, day_analy, i//60, i%60), '%Y-%m-%d %H:%M:%S')
    else:
      dose_time = datetime.strptime("2024-{}-{} {}:{}:00".format(month_analy, day_analy+1, (i-1440)//60, (i-1440)%60), '%Y-%m-%d %H:%M:%S')
    add_data = pd.DataFrame.from_dict([
        {'Time': dose_time,
        'Dose': total_dose(dose_time)
        }])
    df_caffeine = pd.concat([df_caffeine, add_data], ignore_index = True)

  #전체 peak 계산
  peak_dose = max(df_caffeine['Dose'].to_list())

  #Peak까지의 시간 계산
  T_max = math.log(ke / 19.8) / (ke - 19.8)

  #개별 peak 계산 함수
  def individual_peak(index):
    drink_time = df_coffee_history.at[index, 'Coffee_time']
    end_time = drink_time + timedelta(hours = T_max)
    peak = total_dose(end_time)
    return peak

  #HR data import
  df_HR = pd.read_csv(person_file_name, encoding='cp949')
  df_HR.columns = ['TIME', 'HR']
  df_HR['TIME'] = pd.to_datetime(df_HR['TIME'])

  #HR data로부터 concentration 계산
  df_HR['TIME'] = df_HR['TIME'].dt.tz_localize(None)

  def calculate_rr_lf_hf(df_segment):
      rr_intervals = 60000 / df_segment['HR'].values  # 밀리초 단위로 RR 간격 계산
      rr_times = np.linspace(0, len(rr_intervals) * 60 * 1000, num=len(rr_intervals))  # 각 RR 간격에 대한 시간 벡터 생성
      freqs = np.linspace(0.01, 0.4, 800)  # Hz 단위로 주파수 범위 설정
      power = lombscargle(rr_times, rr_intervals, freqs * 2 * np.pi, normalize=True)  # Lomb-Scargle 주기도 계산
      lf = np.trapz(power[(freqs >= 0.04) & (freqs <= 0.15)], freqs[(freqs >= 0.04) & (freqs <= 0.15)])  # LF 파워 계산
      hf = np.trapz(power[(freqs >= 0.15) & (freqs <= 0.4)], freqs[(freqs >= 0.15) & (freqs <= 0.4)])  # HF 파워 계산
      lf_hf_ratio = lf / hf if hf != 0 else 0  # LF/HF 비율 계산
      return lf, hf, lf_hf_ratio

  results = []

  #Test라서 시간 단위 바꿔둠 ~> 실제 코드에는 아래 코드 사용!!!!!
  #start_time = datetime(2024, month_analy, day_analy, 0, 0)
  #end_time = datetime(2024, month_analy, day_analy+1, 0, 0)
  #current_time = start_time

  #Test라서 시간 단위 바꿔둠
  start_time = datetime(2024, 3, 2, 0, 0)
  end_time = datetime(2024, 3, 3, 0, 0)
  current_time = start_time

  #30분 간격으로 나눠둠 -> 실제 데이터 쓸 때는 1분이나 2분 정도로 자를 수 있을듯
  while current_time + timedelta(minutes=30) < end_time:
      next_time = current_time + timedelta(minutes = 30)
      df_segment = df_HR[(df_HR['TIME'] >= current_time) & (df_HR['TIME'] < next_time)]
      if not df_segment.empty:
          lf, hf, lf_hf_ratio = calculate_rr_lf_hf(df_segment)
          results.append((current_time, lf_hf_ratio))
      current_time = next_time

  results_df = pd.DataFrame(results, columns=['Time','lf_hf_ratio'])
  Time_df = results_df.loc[:,'Time']
  lf_hf_ratio_df = results_df.loc[:,'lf_hf_ratio']

  concentration_list = []

  n = 0
  while n < len(lf_hf_ratio_df)-1:
    concen_num = (-0.361) * (lf_hf_ratio_df[n+1] - lf_hf_ratio_df[n])
    concentration_list.append(concen_num)
    n = n+1

  concentration_df = pd.DataFrame(concentration_list,columns=['Concentration'])
  df_conc = pd.concat([Time_df,concentration_df],axis=1)

  #Import data에서 breakpoint 계산 전처리
  df_conc.columns = ['Time', 'Concentration']

  #이 부분은 빼도 될지도?
  df_conc['Time'] = pd.to_datetime(df_conc['Time'])

  Hour = df_conc['Time'].dt.hour
  Minute = df_conc['Time'].dt.minute

  X = Minute + Hour * 60
  Y = df_conc['Concentration']

  dict_conc_analy = {
      'Timestamp': X,
      'Concentration': Y
  }

  df_conc_analy = pd.DataFrame(dict_conc_analy)
  df_conc_analy = df_conc_analy[df_conc_analy['Concentration'].notnull()]
  df_conc_analy.drop(df_conc_analy[(df_conc_analy['Concentration'] == 0 )].index, inplace = True)

  #커피 섭취 시점 전처리
  Hour_coffee = df_coffee_history_person['Coffee_time'].dt.hour
  Minute_coffee = df_coffee_history_person['Coffee_time'].dt.minute
  Time_coffee = Minute_coffee + Hour_coffee * 60

  #효과 유무 별 coffee 섭취 phase 기록
  ineffective_index = []
  effective_index = []

  #전후 breakpoint 계산
  for i in range (len(Time_coffee)):
      target_time = Time_coffee[i]

      #커피 간 간격 3시간 이상이면 섭취 후 3시간까지, 3시간 이하면 다음 커피 섭취 전까지
      if i < (len(Time_coffee)-1):
        if Time_coffee[i+1] - Time_coffee[i] > 180:
          df_conc_analy_sub = df_conc_analy.loc[(df_conc_analy.Timestamp >= (target_time - 30)) & (df_conc_analy.Timestamp < (target_time + 180)), ]
        else:
          df_conc_analy_sub = df_conc_analy.loc[(df_conc_analy.Timestamp >= (target_time - 30)) & (df_conc_analy.Timestamp < (Time_coffee[i+1])), ]
      else:
        df_conc_analy_sub = df_conc_analy.loc[(df_conc_analy.Timestamp >= (target_time - 30)) & (df_conc_analy.Timestamp < (target_time + 180)), ]

      list_Timestamp = df_conc_analy_sub.loc[:, 'Timestamp'].to_list()
      list_Concentration = df_conc_analy_sub.loc[:,'Concentration'].to_list()

      #최적 breakpoint 개수 가져오는 함수
      def min_cal(bic_list, value_num):
          for index, value in enumerate(bic_list):
              if value == value_num:
                  bic_list[index] = 9999999999
          global min_bic
          min_bic = min(bic_list)
          min_bic_num = bic_list.index(min_bic)
          return min_bic_num

      #최적 breakpoint 계산
      if not list_Timestamp:
        ineffective_index.append(i)
        continue

      if not list_Concentration:
        ineffective_index.append(i)
        continue

      ms = piecewise_regression.ModelSelection(list_Timestamp, list_Concentration, max_breakpoints = 3)
      ms.models

      bic_list = [d['bic'] for d in ms.model_summaries]
      min_bic_num = min_cal(bic_list, None)
      if min_bic_num == 0:
          min_bic_num = min_cal(bic_list, min_bic)

      if min_bic_num == 0:
        ineffective_index.append(i)
        continue

      #Breakpoint에 따른 regression line 그리기
      pw_fit = piecewise_regression.Fit(list_Timestamp, list_Concentration, n_breakpoints=min_bic_num)
      pw_fit.summary()

      while not pw_fit.best_muggeo:
          min_bic_num = min_cal(bic_list, min_bic)
          if min_bic_num == 0:
            continue
          pw_fit = piecewise_regression.Fit(list_Timestamp, list_Concentration, n_breakpoints=min_bic_num)
          pw_fit.summary()
          if pw_fit.best_muggeo:
            break

      if min_bic_num == 0:
        ineffective_index.append(i)
        continue

      #만든 regression line에 대한 구체적인 수치
      estimates = pw_fit.best_muggeo.best_fit.estimates

      #Coffee 마시는 시점에 대한 index 구하기
      good = 5
      slope_after_coffee = []
      for j in range(min_bic_num+1):
          if j < min_bic_num:
            raw_estimates = estimates["breakpoint{}".format(j+1)]["estimate"]
          else:
            raw_estimates = estimates["breakpoint{}".format(j)]["estimate"]
          slope_estimates = estimates["alpha{}".format(j+1)]["estimate"]
          if j < good:
              if raw_estimates > target_time:
                  coffee_slope = slope_estimates
                  good = j
              else:
                  continue
          else:
              slope_after_coffee.append(slope_estimates)

      if len(slope_after_coffee) == 0:
        ineffective_index.append(i)
        continue

      #Coffee 마시는 시점에 Concentration 악화되다가 이후에 호전되는 경우 / Coffee 마시는 시점에 Concentration 좋아지다가 이후에 더 좋아지는 경우
      if ((coffee_slope > 0) & (min(slope_after_coffee) < 0)) | ((coffee_slope <= 0) & (min(slope_after_coffee) < coffee_slope)):
          effective_index.append(i)

      #나머지
      else:
          ineffective_index.append(i)

  #하루에 대한 toxic_dose / low_dose 계산
  #각 dose에 대한 list 만들기
  low_dose = []
  effective_dose = []
  toxic_dose = []

  print(ineffective_index)

  #부작용 csv 사용
  df_effect_person = df_effect.loc[df_effect.Name == person_name, :]
  df_effect_person.reset_index(inplace = True)
  adverse_effect = int(df_effect_person.at[0, 'Adverse'])
  sleep_time = df_effect_person.at[0,'Sleep_time']
  sleep_adverse_effect = int(df_effect_person.at[0,'Sleep_adverse'])

  #Adverse effect가 있으면 바로 toxic dose로 분류
  if adverse_effect == 1:
      toxic_dose.append(peak_dose * 0.95)

  else:
      #effective index가 있는 경우
      if len(effective_index) != 0:
        for i in range(len(effective_index)):
          effective_peak = individual_peak(effective_index[i])
          effective_dose.append(effective_peak * 0.95)
      #effective index가 없는 경우
      else:
        low_dose.append(peak_dose * 1.1)

  if len(toxic_dose) != 0:
    today_toxic_dose = min(toxic_dose)
  #임의의 숫자 12 넣음
  else:
    today_toxic_dose = 12

  if len(effective_dose) != 0:
    today_effective_dose = min(effective_dose)
  else:
    today_effective_dose = 12

  if len(low_dose) != 0:
    today_low_dose = max(low_dose)
  else:
    today_low_dose = 0

  #추가 기능 -> 수면 관련하여.. 사용자가 잠에 들기 힘들었다고 생각할 때 -> 누울 때 or 자정의 caffeine dose를 sleep_toxic_dose로 설정..
  sleep_toxic_dose = []

  sleep_hour = sleep_time.hour
  sleep_minute = sleep_time.minute

  if sleep_hour >= 12:
    sleep_time = datetime.strptime("2024-{}-{} {}:{}:00".format(month_analy, day_analy, sleep_hour, sleep_minute), '%Y-%m-%d %H:%M:%S')
  else:
    sleep_time = datetime.strptime("2024-{}-{} {}:{}:00".format(month_analy, day_analy+1, sleep_hour, sleep_minute), '%Y-%m-%d %H:%M:%S')
  sleep_dose = total_dose(sleep_time)

  if sleep_adverse_effect == 1:
    sleep_toxic_dose = sleep_dose

  #임의의 숫자 8 넣음
  else:
    sleep_toxic_dose = 8

    ####### 추가 사항: 최대 5일 전까지의 data만 사용할 수 있게? ######### (코딩 필요)

  #누적 toxic_dose/low_dose 계산 -> 어떻게 해야할지?? (전날 계산한 결과를 저장해야함)
  df_collection_past = pd.read_csv('/content/drive/Shareddrives/Algorithm/카페인 데모데이터/Dose collection.csv', encoding='UTF-8')
  df_collection_past.columns = ['Name', 'Date', 'Toxic_dose', 'Effective_dose', 'Low_dose', 'Sleep_toxic_dose']

  df_collection_today = pd.DataFrame.from_dict([
        {'Name': person_name,
        'Date': "2024-{}-{}".format(str(month_analy).zfill(2), str(day_analy).zfill(2)),
        'Toxic_dose' : today_toxic_dose,
        'Effective_dose' : today_effective_dose,
        'Low_dose' : today_low_dose,
        'Sleep_toxic_dose' : sleep_toxic_dose
        }])

  df_collection_individual_past = df_collection_past.loc[df_collection_past.Name == person_name, :]
  df_collection_individual_present = pd.concat([df_collection_individual_past, df_collection_today], ignore_index=True)

  present_toxic_dose = min(df_collection_individual_present['Toxic_dose'].to_list())
  present_effective_dose = min(df_collection_individual_present['Effective_dose'].to_list())
  present_low_dose = sum(df_collection_individual_present['Low_dose'].to_list()) / len(df_collection_individual_present['Low_dose'].to_list())
  present_sleep_toxic_dose = min(df_collection_individual_present['Sleep_toxic_dose'].to_list())

  df_collection_present = pd.concat([df_collection_past, df_collection_today], ignore_index = True)
  df_collection_present.to_csv('/content/drive/Shareddrives/Algorithm/카페인 데모데이터/Dose collection.csv', mode = 'w', index = False)

  #누적된 Dose cutoff 반영하여 카페인 그래프 다시 산출
  plt.rc('font', family='NanumBarunGothic')
  df_caffeine.plot(x='Time', y='Dose')

  #if len(df_collection_individual_present['Toxic_dose'].to_list()) != 0:
  plt.axhline(y=present_toxic_dose, color='red', linewidth=1)

  if present_effective_dose != 12:
    plt.axhline(y=present_effective_dose, color = 'green', linewidth = 1)
  else:
    plt.axhline(y=present_low_dose, color='green', linewidth=1)


  #len(df_collection_individual_present['Sleep_toxic_dose'].to_list()) != 0:
  plt.axhline(y=present_sleep_toxic_dose, color = 'yellow', linewidth=1)

  plt.ylim([0, peak_dose * 1.6])
  mpl.rcParams['axes.unicode_minus'] = False
  plt.title(person_name + '님의 caffeine graph')
  plt.show()


  print(person_name + "의 부작용 농도는 " + str(present_toxic_dose))
  if present_effective_dose != 12:
    print(person_name + "의 유효 농도는 " + str(present_effective_dose))
  else:
    print(person_name + "의 추정 유효 농도는 " + str(present_low_dose))
  print(person_name + "의 수면 부작용 농도는 " + str(present_sleep_toxic_dose))
```
